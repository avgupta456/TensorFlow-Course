Similar to the Hello World Example, predicting housing prices as a function of number of bedrooms. The relation is price = 50,000 + 50,000 * bedrooms. We divide by 100,000 to simplify calculations. Successfully predicts a 7 bedroom house costing 400,000 dollars.

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([1, 2, 3, 4, 5], dtype=float)
ys = np.array([1.0, 1.5, 2.0, 2.5, 3.0], dtype=float)

model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([7.0]))
```
